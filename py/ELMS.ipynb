```python
from tornado.ioloop import PeriodicCallback
from tornado import gen
```

```python
@gen.coroutine
def dbupdater():
    d = requests.get(url)
    df = json_normalize(d.json()[0], max_level=0).T.head()[0].apply(pd.Series)
    print( DB['test'].insert_all(df.to_dict(orient='records')) )
```

```python
periodInS = 20

periodInMilliS = periodInS * 1000

cbdf = PeriodicCallback(dbupdater, periodInMilliS)
cbdf.start()
```
